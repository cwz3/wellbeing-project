```python
import numpy as np
import pandas as pd
from scipy import stats
def intersection(lst1, lst2):
    lst3 = [value for value in lst1 if value in lst2]
    return lst3
```

```python
#import all DFs and selecting countries

columns = ["Country name",
         "Alcoholic Beverages",
         "Animal Products",
         "Animal fats",
         "Fruits - Excluding Wine",
         "Meat",
         "Milk - Excluding Butter",
         "Starchy Roots",
         "Stimulants", 
         "Sugar & Sweeteners", 
         "Vegetables", 
         "Obesity", 
         "Undernourished", 
         "Population"]
df1 = pd.read_csv("food data/Fat_Supply_Quantity_Data.csv")[columns]
df2 = pd.read_csv("food data/Food_Supply_kcal_Data.csv")[columns]
df3 = pd.read_csv("food data/Food_Supply_Quantity_kg_Data.csv")[columns]
df4 = pd.read_csv("food data/Protein_Supply_Quantity_Data.csv")[columns]
df5 = pd.read_excel("who/DataForFigure2.1WHR2021C2.xls")
happ_exerc = pd.read_csv("216 Project/Happiness_Exercise.csv")
physical = pd.read_excel("216 Project/Physical Activity Data.xlsx")
whr = pd.read_excel("216 Project/WHR20_DataForFigure2.1.xls")
mental = pd.read_csv("216 Project/mental health merged.csv")


countries_happ = happ_exerc["Country name"].tolist()
countries_food = df1["Country name"].tolist()
countries_mental = mental["Country name"].tolist()

countries_relevant = intersection(intersection(countries_happ,countries_food),countries_mental)
#print(len(countries_relevant))

mental.head()


```

```python
#Finalize Datasets with countries_relevant


fat = df1.loc[df1["Country name"].isin(countries_relevant)]
kcal = df2.loc[df1["Country name"].isin(countries_relevant)] #**
quantity = df3.loc[df1["Country name"].isin(countries_relevant)] #** 
protein = df4.loc[df1["Country name"].isin(countries_relevant)]
who = df5.loc[df5["Country name"].isin(countries_relevant)]
happ_exerc = happ_exerc.loc[happ_exerc["Country name"].isin(countries_relevant)]
physical = physical.loc[physical["Country name"].isin(countries_relevant)]
whr = whr.loc[whr["Country name"].isin(countries_relevant)]
mental = mental.loc[mental["Country name"].isin(countries_relevant)]

who.rename(columns={"Country name":"Country"}) #rename W.H.O country column


#display(whr)
#display(physical)
#display(happ_exerc) #do we need this? this just has ladder scores
#display(mental)
display(kcal)
```

```python
#Merging datasets
merged_kcal = kcal.merge(mental.merge(happ_exerc,on="Country name"),on="Country name")
merged_quantity = quantity.merge(mental.merge(happ_exerc,on="Country name"),on="Country name")


display(merged_kcal)
display(merged_quantity)

```
